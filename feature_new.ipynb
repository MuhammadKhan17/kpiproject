```python
# find the most correlated top 3 variables to kpi
kpi_id=['Mean_CTR',
 'Mean_CVRS',
 'Mean_CVRBF',
 'Mean_CVRSV']


# correlation check
data2=data.iloc[:,0:25]
data2=data2.drop(columns=['Creative ID','total_spend','total_Impressions'])
data2.head()
```

```python
corrs=data2.corr(method ='pearson')
corrs
# corrs.to_csv('correlation.csv')
```

```python
list(corrs.columns)
```

```python
from numpy.ma.core import argsort

tops=[]
for id in kpi_id:
  corr_list=corrs[id]
  sort_id=np.argsort(corrs[id])[-7:-1]

  top_correlated=data2.columns[sort_id]
  tops=set(top_correlated)-set(kpi_id)
  print("top correlated variables to "+id+":")
  print(tops)
  print()
```

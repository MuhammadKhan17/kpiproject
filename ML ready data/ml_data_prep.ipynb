```python
import pandas as pd
from sklearn.model_selection import train_test_split

final_dummy_data = pd.read_csv('C:/Users/saulc/OneDrive - University of Calgary/Classes/Econ 599 - OneDrive/ML_data/dummy_ml.csv')
final_labeled_social_data = pd.read_csv('C:/Users/saulc/OneDrive - University of Calgary/Classes/Econ 599 - OneDrive/ML_data/labeled_ml.csv')
```

```python
# Dummy data
X = final_dummy_data.iloc[:, 1:55]
y = final_dummy_data.iloc[:, 55:]

# Split data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# Split training set into training and validation
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=123)

X_train.to_csv('C:/Users/saulc/OneDrive - University of Calgary/Classes/Econ 599 - OneDrive/ML_data/Dummy/dummy_X_train.csv', index=False)
y_train.to_csv('C:/Users/saulc/OneDrive - University of Calgary/Classes/Econ 599 - OneDrive/ML_data/Dummy/dummy_y_train.csv', index=False)
X_val.to_csv('C:/Users/saulc/OneDrive - University of Calgary/Classes/Econ 599 - OneDrive/ML_data/Dummy/dummy_X_val.csv', index=False)
y_val.to_csv('C:/Users/saulc/OneDrive - University of Calgary/Classes/Econ 599 - OneDrive/ML_data/Dummy/dummy_y_val.csv', index=False)
X_test.to_csv('C:/Users/saulc/OneDrive - University of Calgary/Classes/Econ 599 - OneDrive/ML_data/Dummy/dummy_X_test.csv', index=False)
y_test.to_csv('C:/Users/saulc/OneDrive - University of Calgary/Classes/Econ 599 - OneDrive/ML_data/Dummy/dummy_y_test.csv', index=False)
```

```python
# Labeled data
X = final_labeled_social_data.iloc[:, 1:15]
y = final_labeled_social_data.iloc[:, 15:]

# Split data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# Split training set into training and validation
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=123)

X_train.to_csv('C:/Users/saulc/OneDrive - University of Calgary/Classes/Econ 599 - OneDrive/ML_data/Labeled/labeled_X_train.csv', index=False)
y_train.to_csv('C:/Users/saulc/OneDrive - University of Calgary/Classes/Econ 599 - OneDrive/ML_data/Labeled/labeled_y_train.csv', index=False)
X_val.to_csv('C:/Users/saulc/OneDrive - University of Calgary/Classes/Econ 599 - OneDrive/ML_data/Labeled/labeled_X_val.csv', index=False)
y_val.to_csv('C:/Users/saulc/OneDrive - University of Calgary/Classes/Econ 599 - OneDrive/ML_data/Labeled/labeled_y_val.csv', index=False)
X_test.to_csv('C:/Users/saulc/OneDrive - University of Calgary/Classes/Econ 599 - OneDrive/ML_data/Labeled/labeled_X_test.csv', index=False)
y_test.to_csv('C:/Users/saulc/OneDrive - University of Calgary/Classes/Econ 599 - OneDrive/ML_data/Labeled/labeled_y_test.csv', index=False)
```

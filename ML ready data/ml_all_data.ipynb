# ML Preperation: Data Cleaning

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings

# Set graph style
sns.set()

pd.options.display.max_columns = None
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
# Load data
data = pd.read_csv('C:/Users/saulc/OneDrive - University of Calgary/Classes/Econ 599 - OneDrive/KPI Project Data 0120_1221.csv')
labeled_data = pd.read_csv('C:/Users/saulc/OneDrive - University of Calgary/Classes/Econ 599 - OneDrive/labeled_data.csv')
```

## Outlier Detection

```python
week_data = labeled_data.iloc[:, 33:]
week_data.head()
```

```python
week_data.info()
```

```python
ctr_w = [f'CTR_W{i}' for i in range(48)]
cvrbf_w = [f'CVR(BF)_W{i}' for i in range(48)]
cvrsv_w = [f'CVR(SV)_W{i}' for i in range(48)]
cvrint_w = [f'Internet_CVR(S)_W{i}' for i in range(48)]

# Quick look at distributions
fig, ax = plt.subplots(2, 2, figsize=(18, 10), sharex=True)

week_data.reset_index()[ctr_w].plot(ax=ax[0,0], legend=None)
ax[0,0].title.set_text('CTR Distribution')

week_data.reset_index()[cvrbf_w].plot(ax=ax[0,1], legend=None)
ax[0,1].title.set_text('CVR(BF) Distribution')

week_data.reset_index()[cvrsv_w].plot(ax=ax[1,0], legend=None)
ax[1,0].title.set_text('CVR(SV) Distribution')

week_data.reset_index()[cvrint_w].plot(ax=ax[1,1], legend=None)
ax[1,1].title.set_text('Internet CVR(S) Distribution')

plt.show()
```

```python
week_data.describe()
```

We can see some extreme outliers for each KPI. The table above also shows that the difference between the maximum value and the 75th percentile is large. For this reason, we will remove outliers past the 75th percentile.

## Cleaning

The values selected as the cap for each KPI was to find a good balance in the distributions compared to the first graph.

```python
for col in ctr_w:
    cap = 5
    week_data = week_data[week_data[col] < cap]

for col in cvrsv_w:
    cap = 10
    week_data = week_data[week_data[col] < cap]

for col in cvrbf_w:
    cap = 500
    week_data = week_data[week_data[col] < cap]

for col in cvrint_w:
    cap = 100
    week_data = week_data[week_data[col] < cap]
```

```python
# Quick look at distributions
#week_data.reset_index(drop=True, inplace=True)
fig, ax = plt.subplots(2, 2, figsize=(18, 10), sharex=True)

week_data.reset_index()[ctr_w].plot(ax=ax[0,0], legend=None)
ax[0,0].title.set_text('CTR Distribution')

week_data.reset_index()[cvrbf_w].plot(ax=ax[0,1], legend=None)
ax[0,1].title.set_text('CVR(BF) Distribution')

week_data.reset_index()[cvrsv_w].plot(ax=ax[1,0], legend=None)
ax[1,0].title.set_text('CVR(SV) Distribution')

week_data.reset_index()[cvrint_w].plot(ax=ax[1,1], legend=None)
ax[1,1].title.set_text('Internet CVR(S) Distribution')

plt.show()
```

```python
week_data.shape
```

```python
# Potential features to predict
week_data['Mean_CTR'] = week_data[ctr_w].mean(axis=1)
week_data['Mean_CVRSV'] = week_data[cvrsv_w].mean(axis=1)
week_data['Mean_CVRBF'] = week_data[cvrbf_w].mean(axis=1)
week_data['Mean_CVRS'] = week_data[cvrint_w].mean(axis=1)

week_data['Max_CTR'] = week_data[ctr_w].max(axis=1)
week_data['Max_CVRSV'] = week_data[cvrsv_w].max(axis=1)
week_data['Max_CVRBF'] = week_data[cvrbf_w].max(axis=1)
week_data['Max_CVRS'] = week_data[cvrint_w].max(axis=1)

week_data['Median_CTR'] = week_data[ctr_w].median(axis=1)
week_data['Median_CVRSV'] = week_data[cvrsv_w].median(axis=1)
week_data['Median_CVRBF'] = week_data[cvrbf_w].median(axis=1)
week_data['Median_CVRS'] = week_data[cvrint_w].median(axis=1)

kpi_aggs = ['Mean_CTR','Mean_CVRSV', 'Mean_CVRBF', 'Mean_CVRS',
            'Max_CTR', 'Max_CVRSV', 'Max_CVRBF', 'Max_CVRS',
            'Median_CTR', 'Median_CVRSV', 'Median_CVRBF', 'Median_CVRS']
week_data[kpi_aggs]
```

```python
week_data[['Median_CTR', 'Median_CVRSV', 'Median_CVRBF', 'Median_CVRS']].describe()
```

Most of the values for median KPI are 0 so we can drop these. We will also reorder the columns for better readability.

```python
# Drop medians
week_data.drop(['Median_CTR', 'Median_CVRSV', 'Median_CVRBF', 'Median_CVRS'], axis=1, inplace=True)

# Reorder columns
kpi_aggs = ['Mean_CTR','Mean_CVRSV', 'Mean_CVRBF', 'Mean_CVRS',
            'Max_CTR', 'Max_CVRSV', 'Max_CVRBF', 'Max_CVRS']
week_cols = week_data.drop(kpi_aggs, axis=1).columns.to_list()
order = kpi_aggs + week_cols

week_data = week_data[order]
week_data
```

```python
ml_data = pd.merge(data.iloc[:,:18], week_data, left_index=True, right_index=True)
ml_data
```

```python
ml_data.to_csv('ml_full_data.csv', index=False)
```

```python
import pandas as pd
from lazypredict.Supervised import LazyRegressor

# To load the file(s) onto here, go to the folder icon on the left and click the paper with the arrow
# Then load the file(s) from google drive ML ready data
#labeled_data = pd.read_csv('labeled_ml.csv')
ml_data = pd.read_csv('ml_full_data.csv')
ml_data[['Mean_CTR', 'Mean_CVRSV', 'Mean_CVRBF', 'Mean_CVRS']] = ml_data[['Mean_CTR', 'Mean_CVRSV', 'Mean_CVRBF', 'Mean_CVRS']].fillna(0)
ml_data.head()
```

```python
# X = labeled_data.iloc[:,:15]
# y = labeled_data.iloc[:, 15:]
X = ml_data.iloc[:, 1:18]
y = ml_data.iloc[:, 18:]
X['price_applied'] = X['price'].apply(lambda x: 1 if x != 'No Price' else 0)
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

### Mean KPIs

```python
# CTR
reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
models, predictions = reg.fit(X_train, X_test, y_train['Mean_CTR'], y_test['Mean_CTR'])
models
```

```python
reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
models, predictions = reg.fit(X_train, X_test, y_train['Mean_CVRSV'], y_test['Mean_CVRSV'])
models
```

```python
reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
models, predictions = reg.fit(X_train, X_test, y_train['Mean_CVRBF'], y_test['Mean_CVRBF'])
models
```

```python
reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
models, predictions = reg.fit(X_train, X_test, y_train['Mean_CVRS'], y_test['Mean_CVRS'])
models
```

```python
mean_kpis = [col for col in y_train.columns if col.startswith('Mean')]

for kpi in mean_kpis:
  reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
  models, predictions = reg.fit(X_train, X_test, y_train[kpi], y_test[kpi])
  print(f'\n\nPredicting {kpi}')
  print(f'{models}\n\n')

# Top models: HistGradientBoostingRegressor, LGBMRegressor, RandomForestRegressor
```

### Weekly KPIs

```python
# CTR
ctrs = [col for col in y_train.columns if col.startswith('CTR')]

for ctr in ctrs[:10]:
  reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
  models, predictions = reg.fit(X_train, X_test, y_train[ctr], y_test[ctr])

  print(f'\n\nPredicting {ctr}')
  print(f'{models}\n\n')
```

```python
# CVRBF
cvrsv = [col for col in y_train.columns if col.startswith('CVR(SV)')]

for cvr in cvrsv[:10]:
  reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
  models, predictions = reg.fit(X_train, X_test, y_train[cvr], y_test[cvr])

  print(f'\n\nPredicting {cvr}')
  print(f'{models}\n\n')
```
